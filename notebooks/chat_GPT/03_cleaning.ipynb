```python
import re

def clean_text(text: str):

    text = text.strip()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(r"\S+@\S+", "", text)

    text = re.sub(r"<.*?>", "", text)

    text = re.sub(r"\s+", " ", text)

    if len(text) < 5:
        return ""

    return text
```

```python
tests = [
    "سلام      دنیا",
    "https://google.com",
    "<div>Hello</div>",
    "     ",
    "test@gmail.com",
    "سلام، حال شما چطور است؟"
]

for t in tests:
    print("INPUT :", repr(t))
    print("OUTPUT:", repr(clean_text(t)))
    print("-"*40)
```

```python
import lzma
from pathlib import Path

RAW_DATA = Path("../data/raw/fa.txt.xz")

with lzma.open(RAW_DATA, "rt", encoding="utf-8") as f:

    for i, line in enumerate(f):

        cleaned = clean_text(line)

        if cleaned:
            print(cleaned)

        if i >= 20:
            break
```
