```python
import lzma
import re
from pathlib import Path
from tqdm import tqdm
```

```python
RAW_DATA = Path("../data/raw/fa.txt.xz")

OUTPUT = Path("../data/processed/clean_corpus.txt")
```

```python
def clean_text(text):
    
    text = text.strip()

    return text
```

```python
LIMIT = 5_000_000
```

```python
with lzma.open(RAW_DATA, "rt", encoding="utf-8") as infile:

    with open(OUTPUT, "w", encoding="utf-8") as outfile:

        for i, line in enumerate(tqdm(infile)):

            if i >= LIMIT:
                break

            line = clean_text(line)

            if line:
                outfile.write(line + "\n")
```
