```python
import lzma
import re
from pathlib import Path
from tqdm import tqdm
```

```python
RAW_DATA = Path("../data/raw/fa.txt.xz")
OUTPUT = Path("../data/processed/clean_corpus.txt")

LIMIT = 5_000_000
```

```python
import re

def clean_text(text: str):

    text = text.strip()

    # حذف URL
    text = re.sub(r"http\S+", "", text)

    # حذف ایمیل
    text = re.sub(r"\S+@\S+", "", text)

    # حذف HTML
    text = re.sub(r"<.*?>", "", text)

    # حذف فاصله‌های اضافی
    text = re.sub(r"\s+", " ", text)

    # حذف خطوط خیلی کوتاه
    if len(text) < 5:
        return ""

    return text
```

```python
total_lines = 0
kept_lines = 0
removed_lines = 0

with lzma.open(RAW_DATA, "rt", encoding="utf-8") as infile, \
     open(OUTPUT, "w", encoding="utf-8") as outfile:

    for line in tqdm(infile, total=LIMIT):

        if total_lines >= LIMIT:
            break

        total_lines += 1

        cleaned = clean_text(line)

        if cleaned:

            outfile.write(cleaned + "\n")

            kept_lines += 1

        else:

            removed_lines += 1
```

```python
print("=" * 50)

print(f"Total lines   : {total_lines:,}")

print(f"Kept lines    : {kept_lines:,}")

print(f"Removed lines : {removed_lines:,}")

print(f"Output file   : {OUTPUT}")

print("=" * 50)
```
