```python
sp = spm.SentencePieceProcessor()

sp.load(str(config.tokenizer_path))
```

```python
tokens = []

with open(
    config.train_file,
    "r",
    encoding="utf-8",
) as f:

    for line in tqdm(f):

        line = line.strip()

        if not line:
            continue

        ids = sp.encode(line)

        ids.append(sp.eos_id())

        tokens.extend(ids)
```

```python
tokens = np.array(
    tokens,
    dtype=np.uint16,
)

print(tokens.shape)

print(tokens.dtype)
```

```python
dataset = PersianDataset(
    tokens,
    config.max_seq_len,
)

print(dataset)
```
